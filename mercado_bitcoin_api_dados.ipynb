```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_day_summary = pd.DataFrame([text_json])
```

```python
df_day_summary
```

```python
dfs = []
for month in range(1,13):
    month = str(month).zfill(2)
    for day in range(1,32):
        day = str(day).zfill(2)
        date = f'2019/{month}/{day}'
        day_summary = f'https://www.mercadobitcoin.net/api/{coin}/day-summary/{date}/'
        response = requests.get(day_summary)
        if (response.status_code == 200):
            text_json = response.text
            text_json = json.loads(text_json)
            dfs.append(pd.DataFrame([text_json]))
        else:
            print('None')
```

```python
dfs = pd.concat(dfs)
```

```python
dfs = dfs[~dfs.amount.isnull()]
```

```python
df_day_summary = pd.concat([df_day_summary, dfs])
```
